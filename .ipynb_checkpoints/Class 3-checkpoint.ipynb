# Recursion

```python
def fact(n):
    if n == 0:           # base case or exit condition
        return 1
    return n*fact(n-1)   # our recursive case

fact(10)
```

```python
from functools import lru_cache

@lru_cache
def fib(n):
    if n < 2:
        return n
    return fib(n-1) + fib(n-2)

fib(100)
```
